```python
import pandas
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
dataset = load_iris()

x = pandas.DataFrame(dataset['data'], columns=dataset['feature_names'])
y = pandas.Series(dataset['target'], name='target')

x.info()
```

```python
plt.figure(4, figsize=(10, 8))
plt.scatter(x['sepal length (cm)'][:50], x['sepal width (cm)'][:50], color='red', marker='o', label='Iris Setosa')
plt.scatter(x['sepal length (cm)'][50:100], x['sepal width (cm)'][50:100], color='blue', marker='x', label='Iris Versicolor')
plt.scatter(x['sepal length (cm)'][100:], x['sepal width (cm)'][100:], color='green', marker='s', label='Iris Virginica')
plt.xlabel('sepal length (cm)', fontsize=14)
plt.ylabel('sepal width (cm)', fontsize=14)
plt.title('Sepal length VS Sepal width', fontsize=20)
plt.legend(prop = { 'size': 18})
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```
